```python
import numpy as np
import pandas as pd
```

```python
filename = '2020.10.07.002.csv'
data = pd.read_csv(filename, delimiter = ',')

data
```

```python
convergedRows = data['convergenceStatus'] == 'CONVERGED'
data[convergedRows]
```

```python
orig = list(convergedRows)
rows = list(convergedRows)

for i in range(0, len(convergedRows) - 1):
    if convergedRows[i] == True and convergedRows[i+1] == False:
        rows[i] = True
    else:
        rows[i] = False

#rows
data[rows]
```

```python
data[rows].to_csv('points_'+filename)
```
